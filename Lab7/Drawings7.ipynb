```python
import sys
import scipy as sci
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Zad 1
delta = 0.01
x1 = np.arange(0.7, 1.2, delta)
x2 = np.arange(0.7, 1.2, delta)

X1, X2 = np.meshgrid(x1, x2)
Z = 100*((X2 - X1**2))**2 + (1 - X1)**2

fig, ax = plt.subplots()
CS = ax.contour(X1, X2, Z, levels=200)
ax.clabel(CS, inline=False, fontsize=1)

plt.show()
```

```python
#Zad 2A
x1 = np.arange(-2.0, 2.0, delta)
x2 = np.arange(-1, 3.0, delta)

X1, X2 = np.meshgrid(x1, x2)
Z = 2*X1**2 - X1*X2 + 0.5*X2**2 - 3*X1 + 3.5

fig, ax = plt.subplots()
CS = ax.contour(X1, X2, Z, levels=25)
ax.clabel(CS, inline=True, fontsize=10)

plt.show()
```

```python
#Zad2 B
x1 = np.arange(-2.0, 2.0, delta)
x2 = np.arange(-1, 3.0, delta)

X1, X2 = np.meshgrid(x1, x2)
Z = (-1.5)*X1**2 + X1*X2 - 0.5*X2**2 + 2*X1 - 1

fig, ax = plt.subplots()
CS = ax.contour(X1, X2, Z, levels=25)
ax.clabel(CS, inline=True, fontsize=10)

plt.show()
```
