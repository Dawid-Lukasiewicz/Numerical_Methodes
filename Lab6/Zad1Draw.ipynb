```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.linspace(0, 4, 100)
x2 = -0.5*x1 + 2
x3 = -2*x1 + 6
x4 = x1 + 1

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.plot(x1, x2, color = 'green')
plt.plot(x1, x3, color = 'blue')
plt.plot(x1, x4, color = 'orange')
plt.plot(x1, 0*x1)
plt.plot(0*x1, x1)
ax.fill_between(x1, x2, 0, color = 'pink')
ax.fill_between(x1, x2, x3, color = 'white')
ax.fill_between(x1, x2, x4, color = 'white')
plt.xlim([-6, 6])
plt.ylim([-6, 6])
plt.grid(True)
plt.show()
```
