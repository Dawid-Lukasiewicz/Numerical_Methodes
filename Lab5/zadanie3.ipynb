```python
import sys
sys.path.append("../")
from IPython.display import display, Math
import os
import scipy as sci
import numpy as np
import matrix_handler as mx
from solvers_lab5 import Landweber
from solvers_lab5 import Jacobi_iterative
from solvers_lab5 import Gauss_Seidel_iterative
from solvers_lab5 import SOR_method
from solvers_lab5 import SD_method
from solvers_lab5 import Kaczmarz_algorithm
from solvers_lab5 import Grand_Solverr
from solvers_lab5 import solve_error

import matplotlib.pyplot as plt

zadania = "zadania"
```

```python
A = np.loadtxt(os.path.join(zadania, "zad3_A.txt") , float, delimiter=" ", ndmin=2)
b = np.loadtxt(os.path.join(zadania, "zad3_b.txt"), float, delimiter=" ", ndmin=1)
xe = np.loadtxt(os.path.join(zadania, "zad3_xe.txt"), float, delimiter=" ", ndmin=1)
x0 = np.array([0, 0, 0, 0])
```

```python
algorithmss = [Jacobi_iterative, Gauss_Seidel_iterative, SOR_method, SD_method, Kaczmarz_algorithm, Landweber]
xs = Grand_Solverr(A, b, x0, xe, algorithmss)
```

```python
from solvers_lab1 import gauss_elimination_partial_pivot
from solvers_lab1 import back_substitution

Ab = np.loadtxt(os.path.join(zadania, "zad3.txt"), float, delimiter=" ", ndmin=2)
N = len(Ab)
display(Math("[A|b] ="))
mx.print_matrix(Ab)
gauss_elimination_partial_pivot(Ab, N)
x = back_substitution(Ab, N)
display(Math("x ="))
mx.print_matrix(x)

xs.append(x)
```

```python
plt.figure(3)

xError = []
colors = ["black", "orange"]
markerList = ["^", "v"]
labelList = ["Jacobi", "Gauss"]
for i in range(len(xs)):
    xError.append(solve_error(xs[i], xe))
    plt.plot(0, xError[i], ls='dotted', c=colors[i], marker=markerList[i], lw=5, label=labelList[i])

plt.legend()
plt.ylabel("błąd rozwiązania")
plt.show()
```
